## 1.1 Create a two-dimensional vector and plot it on a graph

```
import numpy as np
import matplotlib.pyplot as plt


def plot_2dvectors(vects_2d, axes):
  X, Y, U, V, C= zip(*vects_2d)
  axes.quiver(X, Y, U, V, scale_units='xy', scale=1, angles='xy', color=C)

origin2d = np.array([0,0])
vec2d = np.array([4,2])

vects_2d = [[*origin2d, *vec2d, 'green']] 

fig, axes = plt.subplots()

axes.set_xlim(-1, 6)
axes.set_ylim(-1,3)

axes.set_title('2D Vector')

plot_2dvectors(vects_2d, axes)

plt.show();
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
from collections import defaultdict
mat = {}
mat['A'] = np.array([[1,2],[3,4],[5,6]])

mat['B'] = np.array([[2, 4, 6]])

mat['C'] = np.array([[9,6,3],[4,7,11]])

mat['D'] = np.array([[1,0,0], [0,1,0],[0,0,1]])

mat['E'] = np.array([[1,3],[5,7]])

rows = defaultdict(list)

# Group the matrices by the number of rows 
# rows[nrows] is a list of matrices with
# nrows rows.
for k,m in mat.items():
  rows[m.shape[0]].append(k)

pmat = {}

# For all matrices
for k1,m1 in mat.items():
  #Detrmine the number of columns
  ncols = m1.shape[1]
  # Find all matrices(k2) with ncols rows 
  # For each such matrix, compute the
  # product and store in the dictionary
  # pmat, with key (k1,k2)
  for k2 in rows[ncols]:
    if k1 != k2:
      m2 = mat[k2]  
      pmat[(k1,k2)] = np.matmul(m1, m2)

idx = 1
tot = len(pmat)
for k,m in pmat.items():
  print(f'[{idx}/{tot}] Matrix{k[0]} * Matrix{k[1]}:\n')
  display(m)
  idx += 1
  print(f'Dimensions = {m.shape[0]}x{m.shape[1]}\n')
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12.0, 11.0], [7.0, 10.0]])

J = np.array([[0,1,2],[7,10,4], [3,2, 0]])

detH = np.linalg.det(H)

detJ = np.linalg.det(J)

print(f'{detH:0.5}, {detJ:0.5}')
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
invH = np.linalg.inv(H)

invJ = np.linalg.inv(J)

np.set_printoptions(precision=5, suppress=True)

display(invH)

display(invJ)
```
